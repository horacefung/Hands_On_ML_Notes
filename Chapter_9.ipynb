### TensorFlow

TensorFlow is a powerful open source software library for numerical computation, particularly well suited and fine-tuned for large-scale ML. Its basic principle is simple: **you first define in Python a graph of computations to perform, and then TensorFlow takes that graph and runs it efficiently using optimized C++ code**.

Aside from using C++, one of the main benefits of TensorFlow is that we can break up the graph into several chunks and run them in parallel across multiple CPUs or GPUs. Naturally, TensorFlow also supports distributed computing, so you can train colossal neural networks on humongous training sets in a reasonable amount of time by splitting the computationl across hundres of servers. It was designed by Google and used in many of their large-scale services like Google Clou Speech, Google Photos and Google Search.

### Key Features of TensorFlow:

1. It provides a very simple Python API called TF.Learn2 which is compatible with sklearn. 
2. It provides another simple API called TF-slim to simplify building, training and evaluating neural networks.
3. There are high-level wrapper APIs like Keras or Pretty Tensor
4. You can create all sorts of computations, including any neural network architecture you can think of.
5. Includes many C++ implementations of ML operations. There is also C++ API to define your onw high-performance operations.
6. Provides several optimization nodes that will automatically compute the gradients of the functions you define (auto-diff)
7. Comes with TensorBoard for easy visualizations

```python
#Lets create our first basic graph
#There are no calculations here!
#This is just a computation graph, even the variables are not initialized
x = tf.Variable(3, name = 'x')
y = tf.Variable(4, name = 'y')
f = x*x*y + y + 2
```

```python
f
```

```python
#Lets create a tf session
#We can create tf sessions in different cores to parallelize
sess = tf.Session()
sess.run(x.initializer) #initialize x
sess.run(y.initializer) #initialize y
result = sess.run(f) #run the function
print(result)
sess.close() #free up resources
```
